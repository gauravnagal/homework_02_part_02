Working with the 2010 Q1 dataset break apart the performance data to a training and validation set where the training set is everything before 2015 and the validation set is everything after.

```python
import pandas as pd
```

```python
NAMES = {
    0: "loan_id",
    1: "date",
    2: "serv_name",
    3: "current_int_rate",
    4: "current_act_upb",
    5: "loan_age",
    8: "mat_dt",
    15: "foreclosure_date"}

col_names = [
    NAMES.get(idx, "%d"% idx) for idx, _ in enumerate(range(31))]

perf_df = pd.read_csv("G:/Stevens_Lecture/2019_Fall/FE690_Machine_Learning_in_Finance/HW2/2010Q1/Performance_2010Q1.txt",
                      header = None, 
                      names = col_names, 
                      sep = "|")
```

```python
# setting up training data before year 2015
trn_data = perf_df['date'].map(lambda x: int(x[-4:]) < 2015)
```

```python
train = perf_df[trn_data]
validate = perf_df[~trn_data]
```
